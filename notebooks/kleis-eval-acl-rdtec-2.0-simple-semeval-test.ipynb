```python
import kleis.resources.dataset as kl
from kleis.config.config import ACLRDTEC20, SEMEVAL2017
```

```python
acl_rd_tec_20 = kl.load_corpus(name=ACLRDTEC20)
semeval2017 = kl.load_corpus(name=SEMEVAL2017)
```

```python
acl_rd_tec_20.load_test(annotator="annotator1")
```

```python
print("ACL-RD-TEC 2.0", acl_rd_tec_20.fullname, len(acl_rd_tec_20.train), len(acl_rd_tec_20.test))
```

```python
semeval2017.load_test()
```

```python
print("SemEval 2017", semeval2017.fullname, len(semeval2017.test))
```

```python
acl_rd_tec_20.load_pos_sequences(filter_min_count=1)
for threshold in sorted(acl_rd_tec_20.filtering_counts(), reverse=True):
    print("+++++", threshold)
    acl_rd_tec_20.training(features_method="simple-posseq", filter_min_count=threshold)
    print("P = %f, R = %f, $F_{1.0}$ = %f" % acl_rd_tec_20.eval(dataset=semeval2017.test))
```

```python
semeval2017.load_pos_sequences(filter_min_count=1)
for threshold in sorted(semeval2017.filtering_counts(), reverse=True):
    print("+++++", threshold)
    semeval2017.training(features_method="simple", filter_min_count=threshold)
    print("P = %f, R = %f, $F_{1.0}$ = %f" % semeval2017.eval(dataset=acl_rd_tec_20.test))
```
